```python
# packages

import streamlit as st
import pandas as pd
import io
import requests
import zipfile
import json
from datetime import datetime
import matplotlib
```

```python
# imd extract

url = 'https://assets.publishing.service.gov.uk/media/5d8b3abded915d0373d3540f/File_1_-_IMD2019_Index_of_Multiple_Deprivation.xlsx'

res = requests.get(url)
content = io.BytesIO(res.content)

# Read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(content, sheet_name='IMD2019')

df = dfs
df['Unique_ID'] = range(1, len(df) + 1)
df.set_index('Unique_ID', inplace=True)

df
```

```python
# charity commission data

url = 'https://ccewuksprdoneregsadata1.blob.core.windows.net/data/json/publicextract.charity.zip'

# Make a request to get the ZIP file
res = requests.get(url)

# Unzip the content
with zipfile.ZipFile(io.BytesIO(res.content), 'r') as zip_ref:
    # Assuming there's only one JSON file in the ZIP, you can extract it
    file_name = zip_ref.namelist()[0]
    json_content = zip_ref.read(file_name)

# Load JSON content into a Pandas DataFrame named df2
json_data = json.loads(json_content)
df2 = pd.json_normalize(json_data)  # Adjust as needed based on the JSON structure
```

```python
df2
```

```python
removals = df2[df2['charity_reporting_status'] == 'Removed']
removals
```

```python
removals = removals.drop(columns=['charity_reporting_status',
 'latest_acc_fin_period_start_date',
 'latest_acc_fin_period_end_date',
 'latest_income',
 'latest_expenditure',
 'charity_contact_address1',
 'charity_contact_address2',
 'charity_contact_address3',
 'charity_contact_address4',
 'charity_contact_address5',
 'charity_contact_postcode',
 'charity_contact_phone',
 'charity_contact_email',
 'charity_contact_web',
 'charity_company_registration_number',
 'charity_insolvent',
 'charity_in_administration',
 'charity_previously_excepted',
 'charity_is_cdf_or_cif',
 'charity_is_cio',
 'cio_is_dissolved',
 'date_cio_dissolution_notice',
 'charity_activities',
 'charity_gift_aid',
 'charity_has_land'  
])
list(removals)
```

```python
# add registration/ removals year

removals['Removal Year'] = pd.to_datetime(removals['date_of_removal']).dt.year
removals['Registration Year'] = pd.to_datetime(removals['date_of_registration']).dt.year
removals
```

```python
# length of registration

removals['Registration Duration (Years and Months)'] = (
    pd.to_datetime(removals['date_of_removal']) - pd.to_datetime(removals['date_of_registration'])
).apply(lambda x: f"{x.days // 365} Years, {(x.days % 365) // 30} Months")
removals
```

```python
removals_per_year = removals['Removal Year'].value_counts().reset_index()
removals_per_year.columns = ['Year', 'Number of Removals']
removals_per_year = removals_per_year.sort_values(by='Year')  # This line sorts the DataFrame by the 'Year' column
removals_per_year
```

```python
import matplotlib.pyplot as plt

# Assuming removals_per_year is the DataFrame you created
# Make sure it contains the 'Year' and 'Number of Removals' columns

# Filter years above 2000
removals_per_year_filtered = removals_per_year[removals_per_year['Year'] > 2000]

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(removals_per_year_filtered['Year'], removals_per_year_filtered['Number of Removals'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Removals')
plt.title('Number of Removals per Year (Above 2000)')
plt.show()

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(removals_per_year_filtered['Year'], removals_per_year_filtered['Number of Removals'], marker='o', color='green')
plt.xlabel('Year')
plt.ylabel('Number of Removals')
plt.title('Number of Removals per Year (Above 2000)')
plt.grid(True)
plt.show()
```
